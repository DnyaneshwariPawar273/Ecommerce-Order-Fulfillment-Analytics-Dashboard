```python
#convert date into datetime datatype
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
df['Delivery_Date'] = pd.to_datetime(df['Delivery_Date'], errors='coerce')
```

```python
#extract month from dates
df['Order_Month']=df['Order_Date'].dt.month_name()
df['Delivery_Month']=df['Delivery_Date'].dt.month_name()
```

```python
#calculating delivery duration in days
df['Delivery_Duration_In-Days'] = (df['Delivery_Date'] - df['Order_Date']).dt.days

```

```python
df.loc[df['Delivery_Duration_In-Days'] > 5, 'Delivery_Status'] = 'Delayed'
df.loc[df['Delivery_Duration_In-Days'] <= 5, 'Delivery_Status'] = 'On-Time'
df
```

```python
df['Is_Delayed'] = (df['Delivery_Status'] == 'Delayed').astype(int)
df['Is_On-Time'] = (df['Delivery_Status'] == 'On-Time').astype(int)
df
```

```python
# confirming that the order date should come first then delivery date
invalid_orders = df[df['Order_Date'] > df['Delivery_Date']]
#from result we can see that no invalid orders.
```
